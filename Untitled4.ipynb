```python
import pandas as pd
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/CunyLaguardiaDataAnalytics/datasets/master/2014-15_To_2016-17_School-_Level_NYC_Regents_Report_For_All_Variables.csv')
```

```python
df["Mean Score"] = pd.to_numeric(df["Mean Score"], errors="coerce")
```

```python
df_clean = df.dropna(subset=["Mean Score"])
```

```python
df_clean["Mean Score"].describe()
```

```python
stuy = df_clean[df_clean["School Name"] == "Stuyvesent High School"]
stuy["Mean Score"].describe()
```

```python
overall_mean = df_clean["Mean Score"].mean()
stuy_mean = stuy["Mean Score"].mean()
print(overall_mean, stuy_mean)
```

```python
import matplotlib.pyplot as plt
labels = ["NYC Average", "Stuyvasant HS"]
values = [overall_mean, stuy_mean]
plt.bar(labels, values)
plt.title("Mean Regents Score Comparison")
plt.ylabel("Mean Score")
plt.show()
```

```python
df_clean["Mean Score"].hist()
plt.axvline(stuy_mean, linestyle="dashed")
plt.title("Distribution of Mean Scores Across NYC")
plt.show()
```

```python
borough_map ={"M": "Manahattan",
              "X": "Bronx",
              "K": "Brooklyn",
              "Q": "Queens",
              "R": "Staten Island"}
df["Borough"] = df["School DBN"].str[2].map(borough_map)

```

```python
df["Mean Score"].describe()
```

```python
borough_avg = df.groupby("Borough")["Mean Score"].mean().sort_values(ascending=False)
borough_avg
```

```python
borough_avg.plot(kind="bar")
plt.title("Average Regents Score By Borough")
plt.ylabel("Mean Score")
plt.xticks(rotation=45)
plt.show()
```

```python
labels = ["NYC Avg", "Manhattan Avg","Stuyvasant"]
vales = [
    df["Mean Score"].mean(),
    df[df["Borough"] == "Manahattan"]["Mean Score"].mean(),
    stuy["Mean Score"].mean()
]
plt.bar(labels, vales)
plt.title("Mean Regents Score Comparison")
plt.ylabel("Mean Score")
plt.show()
```
